```python
# To merge all the tables and remove rows with empty churn_label
# USING method will remove the duplicated columns during left join
query6 = """
    SELECT * FROM account a
    LEFT JOIN account_usage au USING (account_id) 
    LEFT JOIN churn_status cs USING (customer_id)
    LEFT JOIN customer cu USING (customer_id)
    LEFT JOIN city c USING (zip_code)
    WHERE cs.churn_label = 'NO' OR cs.churn_label = 'YES';
"""
train_df = pd.read_sql(query6, database_conn)
train_df
```

## Model 1: XGBoost

```python
# Without any transformation
# XGBoost Base Model

from xgboost import XGBClassifier
from sklearn.model_selection import KFold
import statistics
from sklearn import metrics

features = ['tenure_months','total_monthly_fee','age','num_dependents','internet_type','has_premium_tech_support',
'has_online_security','contract_type','payment_method','married']
# 'total_long_distance_fee' and 'total_monthly_fee' have high correlation with 'tenure_months'
X = train_df[features]
y = train_df['churn_label']

kf = KFold(n_splits=10)

model = XGBClassifier(learning_rate=0.1, random_state=5)

k_fold_accuracy = []
auc_score =[]

for train_index, test_index in kf.split(X):
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    XGB_model = model.fit(X_train, y_train)
    y_pred = XGB_model.predict(X_test)
    y_pred_proba = XGB_model.predict_proba(X_test)[:,1]
    
    accuracy = metrics.accuracy_score(y_test, y_pred)
    k_fold_accuracy.append(accuracy)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)
    auc_score.append(auc)
    
print(k_fold_accuracy)
print('The average K fold accuracy is:', statistics.mean(k_fold_accuracy))
print(auc_score)
print('The average auc_score is:', statistics.mean(auc_score))
```

```python
# Final XGBoost Model

from xgboost import XGBClassifier
from sklearn.model_selection import KFold
import statistics
from sklearn import metrics

features = ['tenure_months','total_monthly_fee','age','num_dependents','contract_type','payment_method',
            'married','zip_code','num_referrals']
# 'total_long_distance_fee' and 'total_monthly_fee' have high correlation with 'tenure_months'

X = train_df[features]
y = train_df['churn_label']

kf = KFold(n_splits=10)

model = XGBClassifier(learning_rate=0.1, random_state=5, max_depth=4, gamma=0.8)

k_fold_accuracy = []
auc_score =[]

for train_index, test_index in kf.split(X):
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    XGB_model = model.fit(X_train, y_train)
    y_pred = XGB_model.predict(X_test)
    y_pred_proba = XGB_model.predict_proba(X_test)[:,1]
    
    accuracy = metrics.accuracy_score(y_test, y_pred)
    k_fold_accuracy.append(accuracy)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)
    auc_score.append(auc)
    
print(k_fold_accuracy)
print('The average K fold accuracy is:', statistics.mean(k_fold_accuracy))
print(auc_score)
print('The average auc_score is:', statistics.mean(auc_score))
```
